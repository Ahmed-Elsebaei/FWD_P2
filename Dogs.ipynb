- dogs breeds should be extracted
#### Define
- the highest percentage is the correct one
#### Code and test

```python
req_clean['max_percentage'] = np.nan
for i in range(len(req_clean.index)):
    req_clean['max_percentage'][i]= max([req_clean['first_prediction_confident'][i],req_clean['second_prediction_confident'][i], req_clean['third_prediction_confident'][i]])
```

```python
req_clean['max_percentage']
```

```python
req_clean['dog_breed'] = np.nan
for i in range(len(req_clean.index)):
    if req_clean['max_percentage'][i] == req_clean['first_prediction_confident'][i]:
        req_clean['dog_breed'][i] = req_clean['first_prediction'][i]
    elif req_clean['max_percentage'][i] == req_clean['second_prediction_confident'][i]:
        req_clean['dog_breed'][i] = req_clean['second_prediction'][i]
    else :
            req_clean['dog_breed'][i] = req_clean['third_prediction'][i]
    
        
```
